# 5.	Distribution of Ratings?

```python
rating_distribution = df['rating'].value_counts().dropna()
plt.figure(figsize=(12, 6))
sns.barplot(x=rating_distribution.index, y=rating_distribution.values)
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Distribution of Ratings in Netflix Dataset")
plt.xticks(rotation=45)
plt.show()
```

# 6Which has the highest rating Tv show or Movies?

```python
type_distribution = df['type'].value_counts()
movie_ratings = df[df['type'] == 'Movie']['rating'].value_counts()
tvshow_ratings = df[df['type'] == 'TV Show']['rating'].value_counts()
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
# Movie Ratings Distribution
axes[0].bar(movie_ratings.index, movie_ratings.values, color='skyblue')
axes[0].set_title("Movie Ratings Distribution")
axes[0].set_xlabel("Rating")
axes[0].set_ylabel("Count")
axes[0].tick_params(axis='x', rotation=90)
# TV Show Ratings Distribution
axes[1].bar(tvshow_ratings.index, tvshow_ratings.values, color='lightcoral')
axes[1].set_title("TV Show Ratings Distribution")
axes[1].set_xlabel("Rating")
axes[1].set_ylabel("Count")
axes[1].tick_params(axis='x', rotation=90)
plt.show()
```

# 7.	Finding the best Month for releasing content?

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['month_added'] = df['date_added'].dt.month

# Count the number of releases per month
monthly_distribution = df['month_added'].value_counts().sort_index()

# Plot the distribution
plt.figure(figsize=(10,5))
plt.bar(monthly_distribution.index, monthly_distribution.values, color='skyblue')
plt.xlabel("Month")
plt.ylabel("Number of Releases")
plt.title("Netflix Content Releases by Month")
plt.xticks(range(1,13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Print the monthly distribution
display(monthly_distribution)
```

# 8.Highest watched genres on Netflix?

```python
genre_counts = df['listed_in'].str.split(', ').explode().value_counts()

# Plot the top 10 genres
plt.figure(figsize=(12, 6))
genre_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Watched Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 9.	Released movie over the years?

```python
movies_df = df[df['type'] == 'Movie']

# Count the number of movies per release year
movies_per_year = movies_df['release_year'].value_counts().sort_index()

# Plot the distribution
plt.figure(figsize=(12, 6))
plt.bar(movies_per_year.index, movies_per_year.values, color='skyblue')
plt.xlabel("Release Year")
plt.ylabel("Number of Movies")
plt.title("Distribution of Movies Released Over the Years on Netflix")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 10.movies made on year basis?

```python
movies_per_year = movies_df["release_year"].value_counts().sort_index()

# Plot the distribution
plt.figure(figsize=(12, 6))
plt.plot(movies_per_year.index, movies_per_year.values, marker='o', linestyle='-', color='b')
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.title("Number of Movies Released Per Year on Netflix")
plt.grid(True)
plt.show()
```

# 12.Show all the movies that were released in year 2000.

```python
movies_2000 = df[(df['type'] == 'Movie') & (df['release_year'] == 2000)]
movies_2000_selected = movies_2000[['title', 'director', 'country', 'rating', 'duration', 'listed_in']]

print(movies_2000_selected)
```

# 13.Show only the title of all TV shows that were released in India only.

# 14.Show top 10 director, who gave the highest number of TV shows & Movies to Netflix?

```python
df_directors = df.dropna(subset=['director'])
top_directors = df_directors['director'].value_counts().head(10)

top_directors_df = top_directors.reset_index()
top_directors_df.columns = ['Director', 'Number of Shows & Movies']

print(top_directors_df)
```

# 15.In how many movies/ tv shows, 'tom Cruise' was cast?
